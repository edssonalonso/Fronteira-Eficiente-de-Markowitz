```python
def alocacao_ativos(dataset, dinheiro_total, seed, pesos):

  if seed != 0:#Isso quer dizer que, se Seed=0, nossa funçao vai gerar numeros altetorios diferentes a cada interação
    np.random.seed(seed)

  dataset = dataset.copy()
  
  if len(pesos) <= 0:
    pesos = np.random.random(len(dataset.columns))#retorna lista de números randomicos
    pesos = pesos/pesos.sum() # Aqui estamos "normalizando" os dados, aqui grantimos
    #que todos os pesos serão menores ou iguais a 1, e que a soma dos pesos deem 1
  
  

  acoes_peso = pd.DataFrame({"Acoes": dataset.columns, "Pesos":pesos*100})


  for i in dataset.columns: # "calculando rendimento diário"
    dataset[i] = dataset[i]/dataset[i][0]

  #Vamos transformar os rendimentos diários em dinheiro, conforme o dinheiro investido
  #em cada empresa (seu peso).
  for i, empresa in enumerate(dataset.columns[0:]):
    dataset[empresa] = dataset[empresa]*pesos[i]*dinheiro_total 

  #Agora vamos criar uma coluna somatório, onde vai somar a rentabilidade diária
  #das ações, ou seja, nessa lógica no primeiro dia a soma deve dar 5.000
  #Para isso vamos usar a função sum(axis=1) aqui, ele vai somar linha por linha 
  # e devolver o somatorio
  dataset['Montante'] = dataset.sum(axis=1)

  #Agora vamos criar uma coluna que vai fazer o Retorno Simples Diário da carteira
  dataset["RS Diario Carteira"] = 0.0 #criando coluna para ser percorrida no laco seguinte
  for i in range(1,len(dataset)):
    dataset["RS Diario Carteira"][i] = ((dataset["Montante"][i] - dataset["Montante"][i-1])/dataset["Montante"][i-1])*100

  

  return dataset, acoes_peso, dataset.iloc[len(dataset)-1]["Montante"]
```

```python
#Repeticoes é o numero de vezes que vamos rodar o laco for que esta no algoritimo
#A variável "sem_risco" nos diz se a variável é ou não é uma carteira de risco
def alocacao_portifolio(dataset, dinheiro_total, sem_risco, repeticoes):
  dataset = dataset.copy()
  dataset_virgem = dataset.copy()

  lista_retorno_esperado = []
  lista_volatividade_esperada = []
  lista_sharpe_ratio = []

  #Dentro desta variavel vamos quardar o melhor valor do sharpe ratio, ou seja
  # vamos guardar o maior valor do sharpe ratio que surgir. Vamos dar incialmente
  # o menor valor posiveil que o SO pode dar
  melhor_sharpe_ratio = (1-sys.maxsize)

  #variavel que vai quardar o melhor conjunto de peso (maior sharpe ratio)
  #inicialmente ele comeca vazio, ele comeca sendo uma variavel do tipo numpy
  melhores_pesos = np.empty

  melhor_volatividade = 0
  melhor_retorno = 0

  for rep in range(repeticoes):# Estamos fazendo com que o laco for faca "repeticoes"
     #Gerando pessos aleatorios  
    pesos = np.random.random(len(dataset.columns))
    pesos = pesos/pesos.sum()

    #"Normalizando" os dados
    for i in dataset.columns:
      dataset[i] = dataset[i]/dataset[i][0]

    #transformando o rendimento diário em Dinheiro com os pesos
    for i, empresa in enumerate(dataset.columns):
      dataset[empresa] = dataset[empresa]*pesos[i]*dinheiro_total

    retorno_carteira_log = np.log(dataset/dataset.shift(1))
    matriz_covarianca = retorno_carteira_log .cov()
    #vamos deixar guardado nesta variavel o datatset mas com os retornos logaritmicos
    #diarios de todoas as ações
    dataset["Montante"] = dataset.sum(axis=1)
    dataset["Taxa Retorno Log"] = 0.0
    

    #Soma linha a linha será o montante de cada respectivo dia
    

    #Aqui é a única diferença com a funçao anterior, pois na alocaçao de Markowitz
    #Usamos o Retorno Diário Logaritmico, vamos usar ele para ver o retorno diário
    #da carteira toda
   

    for i in range(1, len(dataset)):
      dataset["Taxa Retorno Log"][i] = np.log(dataset["Montante"][i]/dataset["Montante"][i-1])


    #sharpe_ratio = ( (dataset["Taxa Retorno Log"].mean() - sem_risco)/dataset["Taxa Retorno Log"].std()*np.sqrt(246) )

    retorno_esperado = np.sum(dataset["Taxa Retorno Log"].mean()*pesos)*246
    volatividade_esperada = np.sqrt(np.dot(pesos, np.dot(matriz_covarianca*246, pesos)))

    sharpe_ratio = (retorno_esperado - sem_risco)/volatividade_esperada
    #Aqui retornamos para o dataset original/virgem, pois vamos rodar de novo o laco
    #e nesse novo laco precisamos do dataset virgem
    
    if sharpe_ratio > melhor_sharpe_ratio:
      melhor_sharpe_ratio = sharpe_ratio
      melhores_pesos = pesos
      melhor_volatividade = volatividade_esperada
      melhor_retorno = retorno_esperado
     

    lista_retorno_esperado.append(retorno_esperado)
    lista_sharpe_ratio.append(sharpe_ratio)
    lista_volatividade_esperada.append(volatividade_esperada)

    dataset = dataset_virgem.copy()  

  return melhor_sharpe_ratio, melhores_pesos, lista_volatividade_esperada, lista_retorno_esperado, lista_sharpe_ratio, melhor_retorno, melhor_volatividade
```

```python
#Repeticoes é o numero de vezes que vamos rodar o laco for que esta no algoritimo
#Em cada iteração do laco criaremos uma carteira randomica
def alocacao_portifolio(dataset, dinheiro_total,repeticoes):
  dataset = dataset.copy()
  dataset_virgem = dataset.copy()

  lista_retorno_esperado = []
  lista_volatividade_esperada = []
  lista_sharpe_ratio = []

  for rep in range(repeticoes):
    # Estamos fazendo com que o laco for faca "repeticoes"
    #Gerando pessos aleatorios  
    pesos = np.random.random(len(dataset.columns))
    pesos = pesos/pesos.sum()#"Normalizando" os dados

    #transformacao do dataframe para um que tenha o 
    #rendimento diario em relacao ao preco inicial
    for i in dataset.columns:
      dataset[i] = dataset[i]/dataset[i][0]

    #transformando o rendimento diário em Dinheiro com os pesos
    for i, empresa in enumerate(dataset.columns):
      dataset[empresa] = dataset[empresa]*pesos[i]*dinheiro_total

    retorno_carteira_log = np.log(dataset/dataset.shift(1))
    #vamos deixar guardado nesta variavel o datatset mas com os retornos logaritmicos
    matriz_covarianca = retorno_carteira_log .cov()
    #Matriz de covcariancia calculado sobre o dataframe de retorno log

    dataset["Montante"] = dataset.sum(axis=1)
    dataset["Taxa Retorno Log"] = 0.0

    for i in range(1, len(dataset)):
      dataset["Taxa Retorno Log"][i] = np.log(dataset["Montante"][i]/dataset["Montante"][i-1])

    retorno_esperado = np.sum(dataset["Taxa Retorno Log"].mean()*pesos)*246 
    #calculo do retorno esperado da carteira
    volatividade_esperada = np.sqrt(np.dot(pesos, np.dot(matriz_covarianca*246, pesos)))
    #calculo do risco/desvio padrao da carteira

    lista_retorno_esperado.append(retorno_esperado)
    lista_volatividade_esperada.append(volatividade_esperada)

    dataset = dataset_virgem.copy()
    #Aqui retornamos para o dataset original/virgem, pois vamos rodar de novo o laco

  return  lista_volatividade_esperada, lista_retorno_esperado  
```
